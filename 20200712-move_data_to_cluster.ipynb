```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import gcsfs
import json
import h5py
import os
```

```python
fs = gcsfs.GCSFileSystem(project='learning-2-learn-221016', token="cache")
```

```python
fileNames_old = np.load('data_list_pre_30-70.npy')
```

```python
def get_fileName(oldName, dataName, addition):
    folderName = oldName.split('/chan-')[0]
    preprocess = oldName.split('/chan-')[1].split('/', 1)[1].rsplit('/', 1)[0]
    preprocess_2 = oldName.rsplit('/', 1)[1].split('_chan-')[0]
    preprocess_3 = oldName.split('_chan-')[1].split('_', 1)[1].split('.')[0]
    
    fileName = folderName + \
    '/chan-' + dataName + \
    '/' + preprocess + '/' + addition + '/' + \
    preprocess_2 + \
    '_chan-' + dataName + \
    '_' + preprocess_3 + '_' + addition
    
    return(fileName)
```

```python
def create_h5py_file(data, col_name, filename='new_file'):
    """
    Creates h5py file with data to store

    Parameters
    ----------
    data: the data to be saved.
    col_name: the kind of data to be saved
    filename: the filename in local storage

    Returns
    -------
    filename of local storage
    """
    file = h5py.File(filename, 'w')
    file.create_dataset(col_name, data=data)
    file.close()
    return filename
```

```python
def create_json_file(metaData, filename='metaData.json'):
    """
    Creates json file with premade metadata

    Parameters
    ----------
    metaData: the metadata to be stored
    filename: the filename in local storage

    Returns
    -------
    filename of local storage
    """
    metaFile = filename
    with open(metaFile, 'w') as ff:
        json.dump(metaData, ff)
    return metaFile
```

```python
def save_files(signal, oldName, chan, channels, addition, num_segments, num_ics, d_reduction, fs, overwrite=True):
    fileName = get_fileName(oldName, chan, addition)
    new_loc = fileName + '.mat'
    new_json_loc = fileName + '.json'
    
    if not (fs.exists(new_loc) and fs.exists(new_json_loc) and not overwrite):
        
        file = create_h5py_file(signal, addition, 'z-data-'+chan)
        
        metaData = {'algorithm': 'ICA-picard', 
                    'num_segments': num_segments, 
                    'num_ICS': num_ics,
                    'd_reduction': d_reduction,
                    'channels': str(list(channels))}
        metaFile = create_json_file(metaData, 'z-meta-'+chan)
        
        fs.put(file, new_loc)
        fs.put(metaFile, new_json_loc)
        
        os.remove(file)
        os.remove(metaFile)
        
        returnString = 'File Saved: ' + new_loc.split('.')[0]
        return returnString
    returnString = 'File Existed: ' + new_loc.split('.')[0]
    return returnString
```

```python
reconst_1 = np.load('reconst_1.npy')
channels_1 = np.load('dataNames_1.npy')
```

```python
reconst_2 = np.load('reconst_2.npy')
channels_2 = np.load('dataNames_2.npy')
```

```python
results = []
for i in range(len(channels_1)):
    print(i)
    results.append(save_files(reconst_1[i], fileNames_old[0], channels_1[i], channels_1, 'IC-2', str(80), str(4), 'False', fs, overwrite=False))
```

```python
results = []
for i in range(len(channels_2)):
    print(i)
    results.append(save_files(reconst_2[i], fileNames_old[0], channels_2[i], channels_2, 'IC-2', str(80), str(4), 'False', fs))
```
