```python
import numpy as np
import matplotlib.pyplot as plt
from mne.io import RawArray
from mne import create_info
```

```python
%matplotlib notebook
```

```python
%matplotlib inline
```

```python
data = np.load('resampled.npy')
```

```python
info_m = create_info(data.shape[0], 200, 'misc')
```

```python
raw_m = RawArray(data, info_m)
```

```python
fig = raw_m.plot(n_channels=20)
```

```python
bad_channels = np.array([33,54,67,71,93,113])
```

```python
info = create_info(data[np.array([i for i in np.arange(0,124) if i not in bad_channels])].shape[0], 200, 'misc')
```

```python
raw = RawArray(data[np.array([i for i in np.arange(0,124) if i not in bad_channels])],info)
```

```python
fig = raw.plot(n_channels=20)
```

```python
data_g = data[np.array([i for i in np.arange(0,124) if i not in bad_channels])]
```

```python
data_mean = np.mean(data_g, axis=0)
```

```python
np.arange(0,51,50)
```

```python
matrix = []
for slide in np.arange(-50,51,10):
    matrix.append(np.dot(data_g, np.concatenate((data_mean[slide:], data_mean[:slide]))))
```

```python
plt.imshow(matrix, aspect='auto')
plt.colorbar()
```
